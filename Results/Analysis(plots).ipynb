```python
#pip install matplotlib
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel("Non-variety_streamers.xlsx")
df2 = pd.read_excel("Variety_streamers.xlsx")
plt.figure(figsize = (9,6))
x = df["Index"]
y = df["Unique chatters"]
x2 = df2["Index"]
y2 = df2["Unique chatters"]
plt.title("Percentage of unique chatters (per stream) of non-variety vs variety broadcasters")
plt.ylabel("Percentage of unique chatters")
plt.xlabel("Streams")
plt.plot(x, y, linewidth = 3, markersize=12)
plt.plot(x2, y2, linewidth = 3, markersize=12)
ax = plt.gca()
ax.set_ylim([0,50])
ax.set_xlim([1,18])
plt.legend(['Non-variety Streams', 'Variety Streams'])
i, j = np.polyfit(x2, y2, 1)
plt.plot(x2, i*x2 + j, color = "orange", linestyle = "dashed")
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color = "blue", linestyle = "dashed")
plt.savefig('Percentage_of_unique_chatters_(per_stream)_of_non-variety_vs_variety_broadcasters.png')
```

```python
df = pd.read_excel("Non-variety_streamers.xlsx")
df2 = pd.read_excel("Variety_streamers.xlsx")
plt.figure(figsize = (10,8))
plt.title("Neutrality of non-variety vs variety broadcasters")
plt.ylabel("Percentages")
plt.xlabel("Streams")
ax = plt.gca()
ax.set_ylim([70,90])
ax.set_xlim([0,17])
x = df["Neutral"]
x2 = df2["Neutral"]
plt.plot(x, color='#077B88',linewidth = 3, markersize=12)
plt.plot(x2, color='black',linewidth = 3, markersize=12)
x_mean = [np.mean(x)]*len(x)
ax.plot(x_mean, label='Mean', linestyle='--', color='#077B88')
x2_mean = [np.mean(x2)]*len(x2)
ax.plot(x2_mean, label='Mean', linestyle='--',color='black')
plt.legend(['Non-variety', 'Variety', "Mean of Non-variety", "Mean of Variety"])
plt.savefig('Neutrality_of_non-variety_vs_variety_broadcasters.png')
```

```python
df = pd.read_excel("Non-variety_streamers.xlsx")
df2 = pd.read_excel("Variety_streamers.xlsx")
plt.figure(figsize = (10,8))
plt.title("Negativity of non-variety vs variety broadcasters")
plt.ylabel("Percentages")
plt.xlabel("Streams")
ax = plt.gca()
ax.set_ylim([5,25])
ax.set_xlim([0,17])
y = df["Negative"]
y2 = df2["Negative"]
plt.plot(y, color='#077B88',linewidth = 3, markersize=12)
plt.plot(y2, color='black',linewidth = 3, markersize=12)
y_mean = [np.mean(y)]*len(y)
ax.plot(y_mean, label='Mean', linestyle='--', color='#077B88')
y2_mean = [np.mean(y2)]*len(y2)
ax.plot(y2_mean, label='Mean', linestyle='--',color='black')
plt.legend(['Non-variety', 'Variety', "Mean of Non-variety", "Mean of Variety"])
plt.savefig('Negativity_of_non-variety_vs_variety_broadcasters.png')
```

```python
df3 = pd.read_excel("Bahroo_Results.xlsx")
plt.figure(figsize = (10,8))
plt.title("Analysis of a variety broadcaster")
plt.xlabel("Number of games played in one stream")
plt.ylabel("Percentage of unique chatters")
ax = plt.gca()
ax.set_ylim([15,40])
ax.set_xlim([0,4])
games = df3["Games"]
v = df3["Unique chatters"]
plt.bar(games, v, color="#88AED0")
plt.savefig('Analysis_of_a_variety_broadcaster1.png')
```
